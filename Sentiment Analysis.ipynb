```python
import nltk;
import pandas as pd;
```

```python
whole_data = pd.read_csv("tweets.csv")
data = whole_data.iloc[:10980,:]
data
```

```python
from nltk.tokenize import word_tokenize
x = data.text
y = data.airline_sentiment
from nltk.corpus import stopwords
import string
stop_words = stopwords.words('english')
stop_words.extend(list(string.punctuation))
text = [(j,y[i]) for i in range(len(x)) for j in word_tokenize(x[i]) if not(j.lower() in stop_words)]
text
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
text_x = [i for i, j in text]
l = WordNetLemmatizer()
text2 = [l.lemmatize(i.lower()) for i in text_x]
text2
```

```python
text_x = text2
text_y = [j for i,j in text]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(text_x, text_y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(ngram_range=(0,2))
new_x = vec.fit_transform(x_train)
new_x_test = vec.transform(x_test)
```

```python
vec.get_feature_names_out()
```

```python
from sklearn.naive_bayes import MultinomialNB
rf = MultinomialNB()
rf.fit(new_x,y_train)
rf.score(new_x_test, y_test)
```

```python
data2 = whole_data.iloc[10980:,:]
data2
```

```python
x=data2.text
y=[]
x.reset_index(drop=True,inplace=True)
x
```

```python
for i in range(len(x)):
    alpha = []
    for j in word_tokenize(x[i]):
        if not(j.lower() in stop_words):
            alpha.append(j)
    beta = [l.lemmatize(i.lower()) for i in alpha]
    test = vec.transform(beta)
    y.append(rf.predict(test))
```

```python
y_final = []
for i in y:
    pos=0
    neg=0
    neu=0
    for j in i:
        if j=="positive":
            pos+=1
        elif j=="neutral":
            neu+=1
        else:
            neg+=1
    if(pos>=neg and pos>=neu):
        y_final.append("positive")
    elif(neu>=neg and neu>=pos):
        y_final.append("neutral")
    else:
        y_final.append("negative")
     
```
